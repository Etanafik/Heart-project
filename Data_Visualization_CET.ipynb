```python
import pandas as pd  #Imports the pandas library, which is the primary tool for data manipulation and analysis. It's aliased as pd for brevity.
import matplotlib.pyplot as plt   #Imports Matplotlib's plotting module for creating static, interactive, and animated visualizations. It's aliased as plt.
import seaborn as sns  # Imports the Seaborn library, which is built on top of Matplotlib and provides a high-level interface for drawing attractive and informative statistical graphics. It's aliased as sns.

# Define the file name for the uploaded dataset
file_name = "Heart.csv"     #A variable is assigned the name of the file we want to load.
```

```python

# --- 1. Load and Inspect Data ---
try:
    df = pd.read_csv(file_name)     #Loads the CSV file specified by file_name into a pandas DataFrame object, which we name df
    print("Dataset loaded successfully.")
except FileNotFoundError:
    print(f"Error: The file '{file_name}' was not found. Please ensure it is uploaded to the Colab environment.")
    raise

print("\n--- Data Inspection (First 5 Rows) ---")
print(df.head())  #Uses the .head() method to display the first 5 rows of the DataFrame, providing a quick check of the data structure and values.
```

```python
print("\n--- Data Information ---")
df.info()
```

```python
# --- 2. Data Visualization ---
print("\n--- Data Visualization ---")

# a) Target Variable Distribution (HeartDisease)
plt.figure(figsize=(6, 5))  # Creates a new figure (the canvas for the plot) with a specific size (6 inches wide, 5 inches tall).
sns.countplot(x='HeartDisease', data=df)
plt.title("Heart Disease Distribution (0=No, 1=Yes)", fontsize=14)
plt.xlabel("Heart Disease Status", fontsize=12)
plt.ylabel("Count of Patients", fontsize=12)
plt.xticks([0, 1], ['No Disease (0)', 'Disease (1)'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# b) Age Distribution (Continuous Variable)
plt.figure(figsize=(8, 5))
sns.histplot(df['Age'], kde=True, bins=20, color='skyblue')
plt.title("Age Distribution of Patients", fontsize=14)
plt.xlabel("Age", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# c) Cholesterol vs. Heart Disease (Box Plot)
plt.figure(figsize=(8, 5))
sns.boxplot(x='HeartDisease', y='Cholesterol', data=df, palette='Set2')
plt.title("Cholesterol Levels by Heart Disease Status", fontsize=14)
plt.xlabel("Heart Disease Status (0: No, 1: Yes)", fontsize=12)
plt.ylabel("Cholesterol (mg/dl)", fontsize=12)
plt.xticks([0, 1], ['No Disease', 'Disease'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
